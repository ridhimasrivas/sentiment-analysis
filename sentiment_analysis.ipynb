```python
#sentiment analysis
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
```

```python

# Download required resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Added to fix the LookupError
```

```python
# Sample movie reviews
reviews = [
    "The movie was very interesting and the acting was good",
    "I did not like the film because the story was boring",
    "The direction and screenplay of the movie were excellent"
]
```

```python
# Combine all reviews
text = " ".join(reviews)
```

```python
# Word tokenization (NLTK)
# Note: `punkt` is already downloaded above.
tokens = word_tokenize(text.lower())
print(tokens)

print("\nTotal tokens:", len(tokens))
```

```python
#remove puntuation
tokens = [word for word in tokens if word not in string.punctuation]
print( tokens)
```

```python
# Vocabulary before stop word removal
vocab_before = set(tokens)
```

```python
# Load stop words
stop_words = set(stopwords.words('english'))

# Stop word removal
filtered_tokens = [word for word in tokens if word not in stop_words]

# Vocabulary after stop word removal
vocab_after = set(filtered_tokens)
```

```python
# Results
print("Tokens before stop word removal:\n", tokens)
print("\nVocabulary size before:", len(vocab_before))

print("\nTokens after stop word removal:\n", filtered_tokens)
print("\nVocabulary size after:", len(vocab_after))
```
